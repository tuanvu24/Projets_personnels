```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from  sklearn.model_selection import train_test_split
```

# Prepare the dataset

```python
data = pd.read_csv("/kaggle/input/dataset/AQSONE_Practice_Test_DS_V3_Dataset_A - AQSONE_Practice_Test_DS_V3_Dataset_A.csv")
```

```python
data
```

```python
data.info()
```

```python
# Drop the na value
data.dropna(inplace=True)
data.info()
```

```python
# prepare train, test dataset
X = data.drop(['selling_price'], axis=1)
y = data['selling_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.fuel.value_counts()
```

```python
train_data.name.value_counts()
```

```python
train_data = train_data.drop(['name'], axis=1)
```

```python
train_data.seller_type.value_counts()
```

```python
train_data.transmission.value_counts()
```

```python
train_data.owner.value_counts()
```

```python
train_data.engine.value_counts()
```

```python
train_data = train_data.drop(['engine'], axis=1)
```

```python
# one hot encoding
train_data = train_data.join(pd.get_dummies(train_data.fuel)).drop(['fuel'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data.seller_type)).drop(['seller_type'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data.transmission)).drop(['transmission'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data.owner)).drop(['owner'], axis=1)
```

```python
test_data = X_test.join(y_test)
```

```python
test_data
```

```python
test_data = test_data.join(pd.get_dummies(test_data.fuel)).drop(['fuel'], axis=1)
test_data = test_data.join(pd.get_dummies(test_data.seller_type)).drop(['seller_type'], axis=1)
test_data = test_data.join(pd.get_dummies(test_data.transmission)).drop(['transmission'], axis=1)
test_data = test_data.join(pd.get_dummies(test_data.owner)).drop(['owner'], axis=1)
```

```python
test_data = test_data.drop(['name'], axis=1)
```

```python
test_data = test_data.drop(['engine'], axis=1)
```

# Training model

```python
# random forest
from sklearn.ensemble import RandomForestRegressor

X_train, y_train = train_data.drop(['selling_price'], axis=1), train_data['selling_price']
X_test, y_test = test_data.drop(['selling_price'], axis=1), test_data['selling_price']

model_forest = RandomForestRegressor()
model_forest.fit(X_train, y_train)
model_forest.score(X_test, y_test)
```
